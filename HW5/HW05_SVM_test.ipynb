載入所需要使用的所有函式庫

```python
import cv2
import os
import google.colab.patches as colab
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display
```

Prediction Using SVMs

```python
# 載入訓練好的SVM模型
svm = cv2.ml.SVM_load ("/content/drive/MyDrive/Colab Notebooks/HW5/Train好的模型＿訓練七小時/svm_model.xml");
# 進行測試
```

Setting Label 讀取標籤和圖檔

```python
# 載入txt檔
txt_file = "/content/drive/MyDrive/Colab Notebooks/HW5/test_gt.txt"
# 載入train檔（圖片）
image_floder = "/content/drive/MyDrive/Colab Notebooks/HW5/test"

image_paths = []
labels = []

# 000000.bmp 5
with open(txt_file, 'r') as file: #'r' mean只讀取txt檔
  for line in file:
    parts = line
    image_paths.append(os.path.join(image_floder, parts[0:10])) # parts[0:10] 讀取前十個字元,000000.bmp
    labels.append(parts[11]) # 讀取第11個位元，就是5

# 確認label數量
print(len(labels))

# 測試是否正確
test_image = cv2.imread(image_paths[1])
colab.cv2_imshow(test_image)
print(labels[1])
if labels[1] == '2':
  print("核對正確")
else:
  print("錯誤")
```

**設定HOG**

```python
# 將訓練圖片數據加入，決定HOG特徵提取
winSize =   (28,28) # 視窗大小28x28
blockSize = (28,28) # 28x28
blockStride = (1,1) # 一次讀取一格
cellSize =  (14,14)
nbins = 9
# 創建HOG
HOG = cv2.HOGDescriptor(winSize,blockSize,blockStride,cellSize,nbins)
```

**開始進行測試**

```python
# 設置re_test為需要測試的總數量
re_test = len(labels)
accu_correct, accu_wrong = 0, 0
# 標籤數字從0開始測試
for index in range(re_test):
  in_image = cv2.imread(image_paths[index]) # 隨index計數，讀取圖片
  test_hog = HOG.compute(in_image) # 將圖片轉換成HOG特徵
  test_hog = test_hog.reshape((1,-1)) # 將HOG特徵轉換為 一行 的格式
  test_hog = test_hog.astype(np.float32) # 特徵的資料型態轉換為 float32

  # 使用 SVM模型 進行分類預測
  result = svm.predict(test_hog)
  result = int(result[1])

  if result == int(labels[index]):
    accu_correct = accu_correct + 1
  else:
    accu_wrong = accu_wrong + 1

  print(f"第{index}筆：預測：{result},標籤答案：{labels[index]},累積正確數：{accu_correct},累積錯誤數：{accu_wrong}")
```

**計算準確率**

```python
accuracy = (accu_correct / re_test) * 100
print(f"準確度是：{accuracy}%")
```
