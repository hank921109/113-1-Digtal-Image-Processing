```python
import cv2
import google.colab.patches as colab
import numpy as np
import matplotlib.pyplot as plot
```

**顯示原圖**

```python
in_image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/DIP_hw03_pattern.png",0) #結尾給0，為了讓圖片輸出灰色
colab.cv2_imshow(in_image) #show image
```

```python
#讀取照片的row和col
rows = in_image.shape[0] # shape[0]表示高度
cols = in_image.shape[1] # shape[1]表示寬度

#查看照片像素大小
print(f"Image Height: {rows}")
print(f"Image Width: {cols}")
```

Box Filter算法：九格都*1後/9

```python
box_image = np.zeros((rows,cols),dtype=np.uint8)
for r in range(1,rows - 1):
  for c in range(1,cols - 1):
      box_image[r,c] = (in_image[r-1:r+2,c-1:c+2].sum())/9 #[start:stop-1:start:stop-1:]
```

顯示Box Filter後的圖片

```python
colab.cv2_imshow(box_image) #show image
```

median Filter算法：九格中取中位數

```python
median_image = np.zeros((rows,cols),dtype=np.uint8)
for r in range(1,rows - 1):
  for c in range(1,cols - 1):
      median_image[r,c] = sorted(in_image[r-1:r+2,c-1:c+2].flatten())[4] #取中位數
```

顯示median Filter後的圖片

```python
colab.cv2_imshow(median_image) #show image
```

Laplacian image Enhancement算法：看講義上的九宮格

```python
laplacian_image = np.zeros((rows,cols),dtype=np.uint8)
#array_laplacian = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) # 設定九宮格
#array_laolacian = np.array([[1, 1, 1], [1, -8, 1], [1, 1, 1]]) # 設定九宮格
array_laplacian = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]]) # 設定九宮格
for r in range(1, rows - 1):
    for c in range(1, cols - 1):
        laplacian_image[r, c] = np.sum(in_image[r-1:r+2, c-1:c+2] * array_laplacian) #套入周圍數字相乘
```

顯示Laplacian image Enhancement後的圖片

```python
colab.cv2_imshow(laplacian_image) #show image
```
