```python
in_image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/DIP_hw03_pattern.png",0)
```

```python
colab.cv2_imshow(in_image)
```

get pixel

```python
rows=in_image.shape[0]
cols=in_image.shape[1]
print(f"Image Height: {rows}")
print(f"Image Width: {cols}")
```

box filtering

```python
box_image = np.zeros((rows,cols),dtype=int)
for r in range(rows):
  for c in range(cols):
    if r==0 or r==rows-1 or c==0 or c==cols-1:
      box_image[r,c]=in_image[r,c]
    else:
      box_image[r,c]=(in_image[r-1:r+2,c-1:c+2].sum())/9
colab.cv2_imshow(box_image)
```

median filtering

```python
med_image = np.zeros((rows,cols),dtype=int)
for r in range(rows):
  for c in range(cols):
    if r==0 or r==rows-1 or c==0 or c==cols-1:
      med_image[r,c]=in_image[r,c]
    else:
      temp=sorted(in_image[r-1:r+2,c-1:c+2].flatten())
      med_image[r,c]=temp[4]
colab.cv2_imshow(med_image)
```
