```python
import cv2
import google.colab.patches as colab

in_image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/lena.bmp",0) #結尾給0，為了讓圖片變成灰色
colab.cv2_imshow(in_image) #展示圖片
```

```python
import numpy as np

#讀取照片的row和col
rows = in_image.shape[0] # shape[0]表示高度
cols = in_image.shape[1] # shape[1]表示寬度

#創建upside_down_image
upside_down_image = np.zeros((rows,cols),dtype=np.uint8)

#開始對upside_down_image填寫
for r in range (rows):
  for c in range(cols):
    x = in_image[r,c] # x是變數，用來記錄讀到一格一格的影像
    upside_down_image[rows-1-r,c] = x

#show upside_down_image
colab.cv2_imshow(upside_down_image)
```

```python
#創建right_side_left_image
right_side_left_image = np.zeros((rows,cols),dtype=np.uint8)

#開始對right_side_left_image填寫
for r in range (rows):
  for c in range(cols):
    x = in_image[r,c] # x是變數，表示讀到一格一格的影像
    right_side_left_image[r,cols-1-c] = x

#show right_side_left_image
colab.cv2_imshow(right_side_left_image)
```
